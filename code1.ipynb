```python
from google.colab import files

uploaded = files.upload()  # Upload your audio file (mp3/wav)
audio_path = list(uploaded.keys())[0]
print("Uploaded:", audio_path)
```

```python
import whisper

model = whisper.load_model("base")  # smaller models are faster and stable
result = model.transcribe(audio_path)
print("Full Transcript:\n", result["text"])
```

```python
from pyannote.audio import Pipeline

# Use pre-trained pipeline (requires HuggingFace token)
pipeline = Pipeline.from_pretrained("pyannote/speaker-diarization", use_auth_token="YOUR_HF_TOKEN")

diarization = pipeline(audio_path)

for turn, _, speaker in diarization.itertracks(yield_label=True):
    print(f"{speaker} [{turn.start:.1f}s - {turn.end:.1f}s]:")
```

```python
conversation = []
for turn, _, speaker in diarization.itertracks(yield_label=True):
    speaker_role = "Sales Rep" if speaker == "SPEAKER_00" else "Customer"
    conversation.append(f"{speaker_role} [{turn.start:.1f}-{turn.end:.1f}]: {result['text']}")

for line in conversation:
    print(line)
```
